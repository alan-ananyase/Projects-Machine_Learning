```python
data['Age']=data['Age'].fillna(data['Age'].mean()).astype(int)
data['Salary']=round(data['Salary'].fillna(data['Salary'].mean()),2)
data
```

```python
#sklearn.impute SimpleImputer class works for imputing values in objects or categorical data or object dtype
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='most_frequent')
data['City']=imp.fit_transform(data['City'].values.reshape(-1,1))
data
```

```python
#confirming if all null values are replaced
sns.heatmap(data.isnull())
```

```python
data.columns
```

```python
data.drop(columns=['SNo','Name'],inplace=True)
data
```

```python
print(data['Country'].unique(),'\n',data['City'].unique())
```

```python
#Converting categorical features to numerical features
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data['Country']=le.fit_transform(data['Country'])
data['City']=le.fit_transform(data['City'])
data
```

```python
sns.pairplot(data)
```

```python
data.plot(kind='box',subplots=True,layout=(1,4))
```

```python
y=data.iloc[:,-1]
y.head(5)
```

```python
x=data.drop(columns=['Salary'])
x.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_new=ss.fit_transform(x)
x_new
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_new,y,test_size=0.2,random_state=42)
print('x_train.shape: ',x_train.shape,'x_test.shape: ',x_test.shape,'\ny_train.shape',y_train.shape,'y_test.shape',y_test.shape)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg.fit(x_train,y_train)
y_pred=lreg.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
```

```python
lreg_mae=mean_absolute_error(y_test,y_pred)
lreg_mse=mean_squared_error(y_test,y_pred)
lreg_rmse=np.sqrt(mean_squared_error(y_test,y_pred))
lreg_r2=r2_score(y_test,y_pred)
```

```python
lreg.coef_
```

```python
lreg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
#Ridge Regression (alpha=0.01)
rr001=Ridge(alpha=0.01)
rr001.fit(x_train,y_train)
rr001.score(x_train,y_train)
rr001_y_pred=rr001.predict(x_test)
rr001_mae=mean_absolute_error(y_test,rr001_y_pred)
rr001_mse=mean_squared_error(y_test,rr001_y_pred)
rr001_rmse=np.sqrt(mean_squared_error(y_test,rr001_y_pred))
rr001_r2=r2_score(y_test,rr001_y_pred)
```

```python
#Ridge Regression (alpha=100)
rr100=Ridge(alpha=100)
rr100.fit(x_train,y_train)
rr100.score(x_train,y_train)
rr100_y_pred=rr100.predict(x_test)
rr100_mae=mean_absolute_error(y_test,rr100_y_pred)
rr100_mse=mean_squared_error(y_test,rr100_y_pred)
rr100_rmse=np.sqrt(mean_squared_error(y_test,rr100_y_pred))
rr100_r2=r2_score(y_test,rr100_y_pred)
```

```python
#Lasso Regression (alpha=0.01)
lr001=Lasso(alpha=0.01)
lr001.fit(x_train,y_train)
lr001.score(x_train,y_train)
lr001_y_pred=lr001.predict(x_test)
lr001_mae=mean_absolute_error(y_test,lr001_y_pred)
lr001_mse=mean_squared_error(y_test,lr001_y_pred)
lr001_rmse=np.sqrt(mean_squared_error(y_test,lr001_y_pred))
lr001_r2=r2_score(y_test,lr001_y_pred)
```

```python
#Lasso Regression (alpha=0.0001)
lr00001=Lasso(alpha=0.0001)
lr00001.fit(x_train,y_train)
lr00001.score(x_train,y_train)
lr00001_y_pred=lr00001.predict(x_test)
lr00001_mae=mean_absolute_error(y_test,lr00001_y_pred)
lr00001_mse=mean_squared_error(y_test,lr00001_y_pred)
lr00001_rmse=np.sqrt(mean_squared_error(y_test,lr00001_y_pred))
lr00001_r2=r2_score(y_test,lr00001_y_pred)
```

```python
#ElasticNet is a combination of both Lasso and Ridge
from sklearn.linear_model import ElasticNet
enr=ElasticNet(alpha=0.01)
enr.fit(x_train,y_train)
enr.score(x_train,y_train)
enr_y_pred=enr.predict(x_test)
enr_mae=mean_absolute_error(y_test,enr_y_pred)
enr_mse=mean_squared_error(y_test,enr_y_pred)
enr_rmse=np.sqrt(mean_squared_error(y_test,enr_y_pred))
enr_r2=r2_score(y_test,enr_y_pred)
```

```python
#Support Vector Regression
from sklearn.svm import SVR
score=0
kernel=['linear','poly','rbf']
for i in kernel:
    sv=SVR(kernel=i)
    sv.fit(x_train,y_train)
    a=sv.score(x_train,y_train)
    if a>score:
        score=a
        ker=i
print('Best Kernal: ',ker,'\nScore: ',score)
```

```python
sv=SVR(kernel='poly')
sv.fit(x_train,y_train)
sv.score(x_train,y_train)
sv_y_pred=sv.predict(x_test)
sv_mae=mean_absolute_error(y_test,sv_y_pred)
sv_mse=mean_squared_error(y_test,sv_y_pred)
sv_rmse=np.sqrt(mean_squared_error(y_test,sv_y_pred))
sv_r2=r2_score(y_test,sv_y_pred)
```

```python
final=pd.DataFrame({'Model':['Linear','Lasso 0.01','Lasso 0.0001','Ridge 001','Ridge 100','Elastic Net','Support Vector'],'Score':[lreg.score(x_train,y_train),lr001.score(x_train,y_train),lr00001.score(x_train,y_train),rr001.score(x_train,y_train),rr100.score(x_train,y_train),enr.score(x_train,y_train),sv.score(x_train,y_train)],'Mean Absolute Error':[lreg_mae,lr001_mae,lr00001_mae,rr001_mae,rr100_mae,enr_mae,sv_mae],'Mean Squared Error':[lreg_mse,lr001_mse,lr00001_mse,rr001_mse,rr100_mse,enr_mse,sv_mse]})
final
```
